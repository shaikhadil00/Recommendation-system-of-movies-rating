```python
summary of what each part of the code does:

Loading Data: Reads the ratings data from the CSV file.

Subset Creation: Takes the first 1000 users to make the process manageable.

User-Item Matrix: Creates a matrix where rows represent users and columns represent movies. 
                   Each cell contains the rating a user gave to a movie.

Cosine Similarity: Computes the similarity between users based on their ratings using cosine similarity.

DataFrame: Converts the similarity matrix into a DataFrame for easier manipulation and analysis.
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load the data
ratings = pd.read_csv('ratings.csv')

# Use a manageable subset of the data
ratings_subset = ratings[ratings['userId'] < 1000]  # Taking the first 1000 users

# Create user-item matrix
user_item_matrix = ratings_subset.pivot(index='userId', columns='movieId', values='rating').fillna(0)

# Compute similarity matrix
user_similarity = cosine_similarity(user_item_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)
print(user_similarity_df.head())
```

# Define a function to get top N recommendations:

```python
def get_top_n_recommendations(user_id, n=10):
    # Find similar users
    similar_users = user_similarity_df[user_id].sort_values(ascending=False).index[1:n+1]  # Skip the first user (itself)
    
    # Get the ratings of similar users
    similar_users_ratings = user_item_matrix.loc[similar_users]
    
    # Compute the weighted average of ratings
    weighted_ratings = similar_users_ratings.T.dot(user_similarity_df[user_id].sort_values(ascending=False)[1:n+1])
    
    # Filter out movies already rated by the user
    user_unrated_movies = user_item_matrix.loc[user_id][user_item_matrix.loc[user_id] == 0].index
    
    # Sort and get the top N recommendations
    recommendations = weighted_ratings.loc[user_unrated_movies].sort_values(ascending=False).head(n)
    
    return recommendations.index

# Get top 10 recommendations for user 1
recommended_movies = get_top_n_recommendations(1, 10)
print(f"Top 10 recommendations for user 1: {recommended_movies.tolist()}")
```

# Map Movie IDs to Titles

```python
# Load movie titles
movies = pd.read_csv('movies.csv')

# Get recommended movie titles
recommended_movie_titles = movies[movies['movieId'].isin(recommended_movies)]['title'].tolist()
print(f"Recommended movies for user 1: {recommended_movie_titles}")
```
